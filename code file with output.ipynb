```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import pandas dataFrame
import pandas as pd
```

```python
# load csv file
df=pd.read_csv(r"/content/Dataset .csv")
# shows top five rows from the dataset
df.head()
```

```python
# Display DataFrame info
df.info()
```

```python
# show all columns from the dataset
df.columns.tolist()
```

```python
#provide summary statistics for numerical columns in the DataFrame
df.describe()
```

```python
# Count Total Missing Values in Each Column
print(df.isnull().sum())
```

```python
df1= df.dropna()
df1.shape
```

```python
# Remove duplicate rows
df1.drop_duplicates(inplace=True)
```

```python
columns_to_drop = ['Locality','Locality Verbose','Switch to order menu']
df1.drop(columns=columns_to_drop, inplace=True)

# Check updated columns
df1.columns
```

```python
# Remove rows where 'Average Cost for two', 'Aggregate rating', or 'Votes' contain 0
df1 = df1[(df1['Average Cost for two'] > 0) & (df1['Aggregate rating'] > 0) & (df1['Votes'] > 0)]

# Check dataset shape after removal
print(f"Dataset shape after removing zero values: {df1.shape}")
```

```python
# Count occurrences of each cuisine
top_cuisines = df1['Cuisines'].value_counts().head(3)

# Display the top three cuisines
print(top_cuisines)
```

```python
# Count occurrences of each cuisine
cuisine_counts = df1["Cuisines"].value_counts()

# Calculate percentage
cuisine_percentage = (cuisine_counts / cuisine_counts.sum()) * 100

# Display the top cuisines with their percentage
print(cuisine_percentage.head(3))
print(cuisine_percentage.head(3).apply(lambda x: f"{x:.2f}%"))
```

```python
# Set the style
plt.figure(figsize=(8,5))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette="coolwarm")

# Add labels and title
plt.xlabel("Cuisines", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Top 3 Most Common Cuisines", fontsize=14)
plt.xticks(rotation=45)  # Rotate labels for better readability

# Show the plot
plt.show()
```

```python
# Save the cleaned DataFrame to a new CSV file
df1.to_csv("NEW_CUISINES.csv", index=False)
```
